```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

variables_a_descomponer = [
    'Ventas_semanales',
    'Temperatura',
    'Precio_del_combustible',
    'IPC',
    'Desempleo'
]

PERIODO = 52 # Frecuencia anual para datos semanales

print("Descomposición Clásica para todas las variables...")

for variable in variables_a_descomponer:

    print(f"\nGenerando descomposición para: {variable}")

    # 1. Preparar la serie
    serie_sin_nan = df[variable].dropna()

    # 2. Elegir el modelo: Multiplicativo para Ventas, Aditivo para el resto
    modelo = 'multiplicative' if variable == 'Ventas_semanales' else 'additive'

    # 3. Aplicar la descomposición
    descomposicion_var = seasonal_decompose(
        serie_sin_nan,
        model=modelo,
        period=PERIODO
    )

    # 4. Graficar con el método integrado (solución a la escala del eje Y)
    fig = descomposicion_var.plot()

    # Ajustar el tamaño y el título
    fig.set_size_inches(10, 10)
    fig.suptitle(f'Descomposición Clásica: {variable} (Modelo {modelo})', y=1.02, fontsize=16)

    plt.show()
```

Con lo visto en las graficas se puede observar que se tiene un distinto comportamiento conforme el tiempo, como por ejemplo en las ventas podemos observar unas variaciones fuertes y picos importantes en fechas

en temperatura se observa un comportamiento repetitivo durante el tiempo, lo cual nos muestra una estacionalidad



```python
import numpy as np
import matplotlib.pyplot as plt

# --- Variables a analizar (las 5 series de tiempo continuas) ---
variables_a_analizar = [
    'Ventas_semanales',
    'Temperatura',
    'Precio_del_combustible',
    'IPC',
    'Desempleo'
]

# Parámetros
VENTANA = 50

for variable in variables_a_analizar:

    print(f"\n{'='*70}")
    print(f"VISUALIZACIÓN DE PROMEDIOS MÓVILES: {variable}")
    print(f"{'='*70}")

    serie = df[variable].dropna()

    # --- 1. PROMEDIOS MÓVILES (Cálculo y Gráfico) ---
    # Promedio Móvil Simple (SMA)
    sma = serie.rolling(window=VENTANA).mean()
    # Suavización Exponencial (SES)
    ses = serie.ewm(span=VENTANA, adjust=False).mean()

    plt.figure(figsize=(14, 6))

    # Gráfico de la Serie Original y los Promedios Móviles
    plt.plot(serie, label=f'Serie Original ({variable})', color='purple')
    plt.plot(sma, label=f'Promedio Móvil Simple ({VENTANA} semanas)', color='purple')
    plt.plot(ses, label=f'Suavización Exponencial ({VENTANA} semanas)', color='red')

    plt.title(f'Promedios Móviles de {variable}')
    plt.xlabel('Fecha')
    plt.ylabel(variable)
    plt.legend()
    plt.grid(True)

    # Ajuste de zoom opcional para Ventas Semanales (para ver mejor la suavización)
    if variable == 'Ventas_semanales':
         plt.ylim(1.5e6, 2.5e6)
         print("Nota: Se aplicó un zoom en Y para 'Ventas_semanales' para visualizar mejor la línea suavizada, ya que los picos son muy altos.")

    plt.show()
```

En la temperatura, los promedios móviles muestran un patrón más uniforme y repetitivo, indicando estabilidad en los cambios climáticos. Para el precio del combustible, se observa una tendencia de incremento gradual con algunas variaciones pequeñas.

El IPC mantiene un crecimiento constante a lo largo del tiempo, mientras que el desempleo muestra cambios moderados y relativamente estables. En general, los promedios móviles permiten identificar tendencias más claras y facilitan el análisis de las series de tiempo.

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller
from numpy.polynomial.polynomial import Polynomial

#Variables a analizar
variables_a_analizar = [
   'Ventas_semanales',
    'Temperatura',
    'Precio_del_combustible',
    'IPC',
    'Desempleo'
]

# Parámetros
GRADO_POLINOMIAL = 7       # Usamos el grado 7 para remover la tendencia
NIVEL_SIGNIFICANCIA = 0.05 # Nivel de confianza estándar

for variable in variables_a_analizar:

    print(f"\n{'='*60}")
    print(f"| PRUEBA ADF SOBRE RESIDUALES: {variable}")
    print(f"{'='*60}")

    serie = df[variable].dropna()
    X = np.arange(len(serie))
    y = serie.values

    # Paso 1: Remover la Tendencia Polinomial
    # Esto genera la serie de residuales (serie sin tendencia)
    modelo = Polynomial.fit(X, y, deg=GRADO_POLINOMIAL)
    poly_trend = modelo(X)
    poly_resid = y - poly_trend

    # Paso 2: Realizar la Prueba ADF sobre los residuales
    # Aquí se comprueba la estacionariedad
    adf_test = adfuller(poly_resid)

    print(f"Estadístico ADF: {adf_test[0]:.4f}")
    print(f"p-value: {adf_test[1]:.8f}")

    # Paso 3: Interpretación de la prueba
    if adf_test[1] <= NIVEL_SIGNIFICANCIA:
        print(f"-> La serie de residuales de {variable} es ESTACIONARIA (Rechazamos H₀).")
    else:
        print(f"-> La serie de residuales de {variable} NO es estacionaria (Aceptamos H₀).")
```

```python
#Definición CORREGIDA de la función Transformada_de_Fourier

def Transformada_de_Fourier(serie, terminos, un_grafico, termino):
    '''La función acepta los argumentos "serie" (numpy array), "terminos" (int),
    "un_grafico" (bool) y "termino" (int, índice del componente a guardar).'''

    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    from scipy.fftpack import fft, ifft

    serie_array = serie # Ya es un array de numpy
    n = len(serie_array)
    transformada_fourier = fft(serie_array)
    frecuencias = np.fft.fftfreq(n)

    # Crear DataFrame de componentes
    with np.errstate(divide='ignore', invalid='ignore'):
        longitud_de_onda = 1 / frecuencias
        longitud_de_onda[frecuencias == 0] = np.inf

    componentes = pd.DataFrame({
        'frecuencia': frecuencias,
        'magnitud': np.abs(transformada_fourier),
        'longitud_de_onda': longitud_de_onda
    })
    componentes = componentes.sort_values(by='magnitud', ascending=False)
    # Seleccionar las frecuencias con mayor magnitud (excluyendo la frecuencia cero)
    top_frecuencias = componentes.loc[componentes['frecuencia'] > 0].nlargest(terminos, 'magnitud')
    top_frecuencias.reset_index(drop=True, inplace=True)
    print("Frecuencias principales:\n", top_frecuencias)

    # Preparar Gráficos y Componentes
    plt.figure(figsize=(12, 4))
    componente_temporal_sumado = np.zeros_like(serie_array)
    componentes_temporales = []

    n_loop = 0 # Usamos n_loop en lugar de n, ya que n es la longitud de la serie
    for i, row in top_frecuencias.iterrows():
        n_loop+=1
        fourier_component = np.zeros_like(transformada_fourier)
        idx = np.where(np.isclose(frecuencias, row['frecuencia']))[0][0]

        fourier_component[idx] = transformada_fourier[idx]
        fourier_component[-idx] = transformada_fourier[-idx]

        if n_loop == (terminos + 1):
             break

        componente_temporal = ifft(fourier_component).real
        componentes_temporales.append(componente_temporal)
        componente_temporal_sumado += componente_temporal

        # ELIMINAMOS LA LÍNEA df["seasonal_poly"]
        plt.plot(
            componente_temporal,
            label=f'Longitud de onda {1 / row["frecuencia"]:.0f}',
            alpha=1,
            linewidth = 0.5,
            )
        plt.title('Componentes de Fourier de la Serie')
        plt.xlabel('Tiempo')
        plt.ylabel('Valor')
        plt.legend()
        plt.grid()

    componentes_temporales.append(componente_temporal_sumado)

    # Gráficos finales
    plt.plot(serie_array, label='Serie Original (Residuales)', color='black', alpha=0.5)

    if not un_grafico:
        plt.figure(figsize=(12, 4))
        plt.plot(serie_array, label='Serie Original (Residuales)', color='black', alpha=0.5)

    plt.plot(componente_temporal_sumado, label='Componente temporal sumada', color='green')
    plt.legend()
    plt.title('Suma de los Componentes de Fourier de la Serie')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor')
    plt.grid()
    plt.show()

    return componentes_temporales
```

```python
# Bloque Final: Prueba ADF para las 5 series de tiempo

from statsmodels.tsa.stattools import adfuller

variables_a_analizar = [
    'Ventas_semanales',
    'Temperatura',
    'Precio_del_combustible',
    'IPC',
    'Desempleo'
]
nivel_de_significancia = 0.05

for variable in variables_a_analizar:

    columna_residual = f'Poly_Resid_{variable}'

    if columna_residual in df.columns:

        serie_residual = df[columna_residual].dropna()

        print(f"\n{'='*80}")
        print(f"PRUEBA ADF para los residuales de: {variable}")
        print(f"{'='*80}")

        # Realizar la prueba ADF
        try:
            adf_test = adfuller(serie_residual)

            # 1. Interpretación de la prueba ADF
            print(f"Estadístico ADF: {adf_test[0]:.4f}")
            print(f"p-value: {adf_test[1]:.4f}")
            print("Valores Críticos:", adf_test[4])

            # 2. Conclusión
            if adf_test[1] <= nivel_de_significancia:
                print("Conclusión: La serie de residuales es ESTACIONARIA (rechazamos la hipótesis nula).")
            else:
                print("Conclusión: La serie de residuales NO es estacionaria (aceptamos la hipótesis nula).")

        except Exception as e:
            print(f"Error al realizar la prueba ADF para {variable}: {e}")

    else:
        print(f"ERROR: La columna de residuales '{columna_residual}' no fue encontrada.")
```
